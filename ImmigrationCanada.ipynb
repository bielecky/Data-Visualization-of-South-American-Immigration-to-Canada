```python
import pandas as pd
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('Country', inplace=True)
```

```python
anos = list(map(str,range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brazil', anos]
```

```python
brasil
```

```python
brasil_dict = {'anos': brasil.index.tolist(), 'immigrants': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
brasil_dict
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['anos'], dados_brasil['immigrants'])
plt.title('Immigration from Brazil to Canada')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['anos'], dados_brasil['immigrants'])
ax.set_title('Immigration from Brazil to Canada\nfrom 1980 to 2013')
ax.set_xlabel('Year')
ax.set_ylabel('Number of immigrants')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))

axs[0].plot(dados_brasil['anos'], dados_brasil['immigrants'])
axs[0].set_title('Immigration from Brazil to Canada\n from 1980 to 2013')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Number of immigrants')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['immigrants'])
axs[1].set_title('Boxplot of Immigration from Brazil to Canada\nfrom 1980 to 2013')
axs[1].set_xlabel('Brazil')
axs[1].set_ylabel('Immigrants')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Immigration from the 4 Largest South American Countries to Canada')

axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brazil')

axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

  for ax in axs.flat:
    ax.set_xlabel('Year')
    ax.set_ylabel('Number of immigrants')

    ymin = 0
    ymax = 7000

    for ax in axs.ravel():
      ax.set_ylim(ymin, ymax)

plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y='Country', x='Total', orient='h', palette=palette)
  ax.set_title('Countries with highest immigration to Canada', fontsize=18)
  ax.set_xlabel('Number of immigrants', fontsize=14)
  ax.set_ylabel('')

  plt.show()
```

```python
fig = px.line(dados_brasil, x='anos', y='immigrants',
              title='Immigration from Brazil to Canada')
fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle':-45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Year',
                  yaxis_title='Number of immigrants')
fig.show()
```
